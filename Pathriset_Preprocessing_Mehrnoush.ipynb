#### Probabilistic Imputation for string:Probabilistic imputation for string data involves replacing missing values with the most likely value based on the probability distribution of the non-missing values. One common way to achieve this is by randomly sampling values from the distribution of the non-missing values, weighted by their probabilities.

```python
imput_df.isnull().sum()
```

```python
value_counts = imput_df['employment_status'].value_counts(normalize=True)
```

```python
missing_indices = imput_df[imput_df['employment_status'].isnull()].index
```

```python
missing_imputed = np.random.choice(value_counts.index, size=len(missing_indices), p=value_counts.values)
```

```python
imput_df.loc[missing_indices, 'employment_status'] = missing_imputed
```

```python
z_scores = np.abs(stats.zscore(imput_df['number_of_interviews']))
z_score_threshold = 3
outliers = imput_df[z_scores > z_score_threshold]
print(outliers['number_of_interviews'].count())
```

```python
z_scores = np.abs(stats.zscore(imput_df['number_of_applications']))
z_score_threshold = 9
outliers = imput_df[z_scores > z_score_threshold]
print(outliers['number_of_applications'].count())
```

# Encoding categorical data

This approach is suitable when dealing with ordinal categorical variables where there's a clear order or hierarchy among the categories Such as 'highest_level_of_education' , 'length_of_job_search ','professional_experience' and 'year' at pathrise project.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
Encod_df=dummy_df.copy()
```

```python
LE = LabelEncoder()
for column in ['highest_level_of_education','length_of_job_search','professional_experience','year']:
    Encod_df[column] = LE.fit_transform(Encod_df[column])
```
